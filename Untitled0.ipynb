```python

import statistics as st
import math
import matplotlib.pyplot as plt
from collections import Counter
from scipy.stats import poisson, skew, norm, binom


# TABELA 1 – Notas dos alunos

notas = [68,67,68,78,68,52,73,66,67,71,72,83,77,71,61,57,72,85,70,68,76,52,66,75,80,67,74,72,79,56]

# Média, moda, mediana e desvio padrão
media1 = st.mean(notas)
moda1 = st.mode(notas)
mediana1 = st.median(notas)
desvio1 = st.pstdev(notas)

print("=== TABELA 1 ===")
print("Média:", media1)
print("Moda:", moda1)
print("Mediana:", mediana1)
print("Desvio padrão:", desvio1)

# Probabilidade de nota >= 80
p_80 = len([n for n in notas if n >= 80]) / len(notas)
print("Probabilidade de nota >= 80:", p_80)

# Gráfico e assimetria
plt.hist(notas, bins=10, edgecolor='black')
plt.title("Distribuição das Notas")
plt.xlabel("Notas")
plt.ylabel("Frequência")
plt.show()

assimetria = skew(notas)
print("Assimetria:", assimetria)
if abs(assimetria) < 0.5:
    print("A distribuição é aproximadamente simétrica (Normal).")
else:
    print("A distribuição não é Normal.")

defeitos = [2,1,5,0,2,3,2,1,4,1,2,3,3,0,2,3,1,2,0,2,2,1,2,1,2,2,8,5,3,2]

# Média (λ) e desvio padrão
media2 = st.mean(defeitos)
desvio2 = st.pstdev(defeitos)

print("\n=== TABELA 2 ===")
print("Média (λ):", media2)
print("Desvio padrão:", desvio2)

# P(X = 0)
p0_teorica = poisson.pmf(0, media2)
dias_sem_defeito = defeitos.count(0)
p0_observada = dias_sem_defeito / len(defeitos)

print("P(X=0) teórica:", p0_teorica)
print("P(X=0) observada:", p0_observada)

# P(X >= 5)
p_maior5 = 1 - poisson.cdf(4, media2)
print("P(X ≥ 5):", p_maior5)

# Gráfico
plt.plot(defeitos, marker='o')
plt.title("Defeitos por dia - Janeiro 2024")
plt.xlabel("Dia")
plt.ylabel("Número de defeitos")
plt.show()


conversoes = [7,11,12,11,9,7,11,15,8,11,22,17,12,16,11,17,14,18,21,19]
p1 = 0.06
p2 = 0.09

# 1ª fase (dias 1–10) e 2ª fase (11–20)
fase1 = conversoes[:10]
fase2 = conversoes[10:]

media_f1 = st.mean(fase1)
media_f2 = st.mean(fase2)
print("\n=== TABELA 3 ===")
print("Média Fase 1:", media_f1)
print("Média Fase 2:", media_f2)

# Probabilidade de observar pelo menos a média de conversões
# Supondo n = 100 tentativas por dia (exemplo prático)
n = 100
p_f1 = 1 - binom.cdf(int(media_f1)-1, n, p1)
p_f2 = 1 - binom.cdf(int(media_f2)-1, n, p2)
print("P(X ≥ média Fase 1):", p_f1)
print("P(X ≥ média Fase 2):", p_f2)

# Média e desvio padrão gerais
media3 = st.mean(conversoes)
desvio3 = st.pstdev(conversoes)
print("Média geral:", media3)
print("Desvio padrão geral:", desvio3)

# Gráfico
plt.plot(conversoes, marker='o')
plt.title("Conversões diárias")
plt.xlabel("Dia")
plt.ylabel("Conversões")
plt.show()

# Assimetria
assimetria3 = skew(conversoes)
print("Assimetria:", assimetria3)
if abs(assimetria3) < 0.5:
    print("Distribuição aproximadamente simétrica.")
else:
    print("Distribuição assimétrica.")
```
